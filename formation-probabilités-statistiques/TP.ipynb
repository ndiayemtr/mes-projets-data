```python
# Séparer les colonnes quantitaives(numerique)
num_data = df.select_dtypes(include=['int64', 'float64'])

# Séparer les colonnes qualitaives(categorielles)
cat_data = df.select_dtypes(include=['object'])
```

```python
cat_data = cat_data.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data.isnull().sum().sort_values(ascending=False)
```

```python
num_data.isnull().sum().sort_values(ascending=False)
```

```python
num_data.fillna(method='bfill', inplace=True)
num_data.isnull().sum().sort_values(ascending=False)
```

```python
# Transformer la colonne target (Loan_Status)
target_value = {'Y': 1, 'N':0}
target = cat_data['Loan_Status']
cat_data.drop('Loan_Status', axis=1, inplace=True)
target=target.map(target_value)
target
```

```python
# Transformer les valeurs cat en num
le = LabelEncoder()
for i in cat_data:
  cat_data[i]=le.fit_transform(cat_data[i])
cat_data
```

```python
# Supprimer Loan_ID
cat_data.drop('Loan_ID', axis=1, inplace=True)
```
