Echantillonnage & Biais en Statistique

Definiion
Un Echantillonnage est un sous-ensemble de la population utilisé pour:
* Estimer des parametres
* Modeliser des comportements
* Comprendre de population totale

# Types de sampling (Échantillonnage)

1. **Échantillonnage aléatoire simple**

Chaque ligne de votre dataset a la meme probapilité d'etre selectionnée.

```python
sample_simple = df.sample(n=20, random_state=24)
sample_simple
```

2. Échantillonnage stratifié

On échantilonne proportionnellement selon une variable categorielle, par exemple: **segment_age** ou **revenu_cat**

```python
# Exemple : stratification par segment_age

if 'segment_age' in df.columns:
  sample_stratifie = df.groupby('segment_age').sample(n=5, random_state=42)
else:
  sample_stratifie = df.groupby(df.columns[0]).sample(n=5, random_state=42)

sample_stratifie
```

```python
# Proportion par segment

n = 200
proportions = df['segment_age'].value_counts(normalize=True)
n_par_segment = (proportions * n).round().astype(int)

print(proportions)
print(n_par_segment)
```

```python
# 2. Colonnes sans la colone de regroupement

group_col = 'segment_age'
other_cols = [c for c in df.columns if c != group_col]

# 3. Echantillonnage manuel propre (sans warnig)
sample_manuel = (
      df.groupby(group_col)[other_cols]
      .apply(lambda g: g.sample(n=n_par_segment.loc[g.name], random_state=42))
      .reset_index(level=0)
      .reset_index(drop=True)
)
# 4. Répartition dans l'echantillon
sample_manuel['segment_age'].value_counts(normalize=True)
```

```python
print('Taille population:', len(df))
print(df['segment_age'].value_counts(normalize=True))

print('\nTaille éhantillon:', len(sample_manuel))
print(sample_manuel['segment_age'].value_counts(normalize=True))
```

# 2.2 — Types de biais

un biais est une erreure systématique qui déforme les données ou analyse et qui conduit a des conclusions incorretes.

1. Biais de sélection
Certains clients ont plus de chances d'apparaître dans les données.

**Exemples** e-commerce :

seules les personnes ayant finalisé un achat apparaissent
les clients ayant un panier élevé remplissent mieux les formulaires

🔹 2. Biais du survivant
On n’observe que ceux qui restent.

**Exemples** e-commerce :

analyser uniquement les clients fidèles → les churners disparaissent
regarder seulement les produits à succès → les flops invisibles

🔹 3. Biais d’auto-sélection
Les individus se sélectionnent eux-mêmes.

**Exemples** :

seuls les clients motivés répondent aux questionnaires de satisfaction
les avis produits viennent surtout des extrêmes (très satisfaits / très déçus)
